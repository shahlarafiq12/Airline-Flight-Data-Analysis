```python
#Showing all the Airlines with their number of flights in bar graph
df['airline'].value_counts(ascending = True).plot.barh(color = ['blue'])
plt.title('Airlines Flights Data', fontsize = 20)
plt.xlabel('Number of Flights')
plt.ylabel('Airlines')
plt.show()
```

```python
counts = df['airline'].value_counts(ascending=True)

ax = counts.plot.barh(color='blue', figsize=(8,6))
plt.title('Airlines Flights Data', fontsize=20)
plt.xlabel('Number of Flights')
plt.ylabel('Airlines')

# Add bar labels
ax.bar_label(ax.containers[0], fmt='%d')  # '%d' formats as integer

plt.tight_layout()
plt.show()
```

#Q2. Show bar graphs representing the departure time & arrival time

```python
#showing the departure time count
df['departure_time'].value_counts()
```

```python
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.bar(df['departure_time'].value_counts().index, df['departure_time'].value_counts().values, color = ['blue', 'orange'])
plt.title('Departure Time Bar graph', fontsize=20)
plt.xlabel('D. Time')
plt.ylabel('D. Frequency')

plt.subplot(1,2,2)
plt.bar(df['arrival_time'].value_counts().index, df['arrival_time'].value_counts().values, color = ['blue', 'purple'])
plt.title('Arrival Time Bar graph', fontsize=20)
plt.xlabel('A. Time')
plt.ylabel('A. Frequency')
plt.show()
```

#Q4. Does price varies with airlines?

```python
# Grouping the airlines and checking their mean price

df.groupby('airline')['price'].mean()
```

```python
#Showing a categorical plot showing the mean ticket price for each airline
g = sns.catplot(
    x='airline',
    y='price',
    kind='bar',
    hue='class',
    data=df,
    height=6,
    aspect=1.5
)

# Add bar labels
for ax in g.axes.flat:
    for container in ax.containers:
        ax.bar_label(container, fmt='%.0f', label_type='edge')  # integer labels on top

# plt.tight_layout()
plt.show()
```

#Q5. Does Ticket Price change based on the departure time and arrival time?

```python
#Checking the Mean Ticket Price based on the Departure Times

df.groupby('departure_time')['price'].mean()
```

```python
sns.catplot(x = 'departure_time', y = 'price', kind = 'bar', data = df, hue = 'departure_time')
plt.xticks(rotation = 35)
plt.show()
```

```python
#Checking the mean ticket price based on the arrival time
df.groupby('arrival_time')['price'].mean()
```

```python
sns.catplot(x = 'arrival_time', y = 'price', kind = 'bar', data = df, hue = 'arrival_time')
plt.xticks(rotation = 35)
plt.show()
```

```python
sns.relplot(x = 'arrival_time', y = 'price', data = df, col = 'departure_time', kind = 'line')
plt.show()
```
